## Importar librerías

```python
from openai import OpenAI
from dotenv_vault import load_dotenv
import os
```

## Cargar API Key de OpenAI

```python
# Cargar variables de entorno desde el archivo .env
load_dotenv()

# Obtener la API Key desde las variables de entorno
OPENAI_KEY = os.getenv('OPENAI_KEY')

# Si estás en Google Colab, puedes usar esta alternativa:
# from google.colab import userdata
# OPENAI_KEY = userdata.get('OPENAI_KEY')

openai = OpenAI(api_key=OPENAI_KEY)
```

## Clasificador de Tweets con Completion API (Legacy)

```python
response = openai.responses.create(
    model="gpt-5-nano",
    input="Decide si el sentimiento de un Tweet es positivo, neutral, o negativo. \
    \n\nTweet: \"uno de los comandos que más me han ayudado últimamente a ubicar algo que necesito sin buscar archivo por archivo en software que no conozco del todo, les dejo esto por si les sirve\
    \"\nSentiment:",
    top_p=1.0,
)
```
