```python
from openai import OpenAI
from dotenv_vault import load_dotenv
import os
# Cargar variables de entorno desde el archivo .env
load_dotenv()

# Obtener la API Key desde las variables de entorno
OPENAI_KEY = os.getenv('OPENAI_KEY')

# Si estás en Google Colab, puedes usar esta alternativa:
# from google.colab import userdata
# OPENAI_KEY = userdata.get('OPENAI_KEY')

openai = OpenAI(api_key=OPENAI_KEY)
```

```python
import random

def get_clue():
  words = ['elefante', 'león', 'jirafa', 'hipopótamo', 'mono']
  random_word = random.choice(words)
  prompt = 'Adivina la palabra que estoy pensando. Es un animal que vive en la selva.'
  return prompt, random_word
```

```python
def check_answer(user_input, answer):
  return user_input == answer    
```

```python
def give_property(animal):
  response = openai.responses.create(
      model= 'gpt-5-nano',
      input='Dame una caracteristica del tipo animal ' + animal + ', pero jamás digas el nombre del animal y procura no repetir propiedades del animal que ya has dicho antes.',
      instructions= 'Este es un juego de adivinanzas. El usuario piensa en un animal y tu le das una pista para que adivine en el formato "Pista: <Pista> ". Jamás digas el nombre del animal. Y limitate a dar solamente la respuesta, no digas nada mas',
      reasoning={"effort": "low"},
      )
  return response.output_text
```

```python
def play_game():
  prompt, answer = get_clue()
  print(prompt)
  while True:
    user_input = input('Ingresa tu respuesta:')
    if check_answer(user_input,answer):
      print('Correcto! La respuesta era:', answer)
      break
    else:
      print('Respuesta incorrecta. Intentalo de nuevo')
      print(give_property(answer))
```

```python
play_game()
```
